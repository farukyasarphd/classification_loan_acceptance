## Preprocessing

```python
# Define features and target
features = data.drop(columns=['ID', 'Experience', target])
labels = data[target]
```

```python
# Uncomment to handle missing data and impute with median values. Current data has no missing values.  
# features = features.fillna(features.median())
```

### Split Data

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=35)
```

```python
# Numerical vars
numerical_cols = features.select_dtypes(include=["float64", "int64"]).columns
print("Numerical Columns:", numerical_cols)

# Uncomment if there are categorical vars
#categorical_cols = features.select_dtypes(include=["object"]).columns
#print("Categorical Columns:", categorical_cols)
```

```python
# Preprocessing pipeline - handles numerical columns as is. Uncomment below if there are categorical vars. 
preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline(steps=[("imputer", SimpleImputer(strategy="median")), 
                                ("scaler", StandardScaler())]), numerical_cols)
#        ,("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
    ]
)

# Fit and transform the training data
X_train_transformed = preprocessor.fit_transform(X_train)

# Transform the testing data
X_test_transformed = preprocessor.transform(X_test)

# Display the transformed data shapes
print("Transformed X_train shape:", X_train_transformed.shape)
print("Transformed X_test shape:", X_test_transformed.shape)
```

### Balance Data

I use SMOTE to create synthetic samples for the minority class (in our case, loan acceptance) to address the class imbalance.

```python
# Balance the training set using SMOTE
smote = SMOTE(random_state=35)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_transformed, y_train)

# Check the class distribution after balancing
print("\nOriginal target distribution:")
print(y_train.value_counts())
print("\nBalanced target distribution:")
print(pd.Series(y_train_balanced).value_counts())
```

## Define and Train Models

```python
# Define the models to compare and set the random state
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=35),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=35),
    "Gradient Boosting": GradientBoostingClassifier(random_state=35),
    "XGBoost": XGBClassifier(eval_metric="logloss", random_state=35)
}

# Function to train and evaluate models
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    metrics = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_prob) if y_prob is not None else "N/A"
    }
    cm = confusion_matrix(y_test, y_pred)

    return metrics, cm
```

### Train Models

```python
# Train and evaluate each model
results = {}
confusion_matrices = {}
for model_name, model in models.items():
    print(f"Training and evaluating {model_name}...")
    metrics, cm = evaluate_model(model, X_train_balanced, y_train_balanced, X_test_transformed, y_test)
    results[model_name] = metrics
    confusion_matrices[model_name] = cm
print("\nDone!")
```

### Compare Models 

```python
# Convert DataFrame for comparison
results_df = pd.DataFrame(results).T
print("\nModel Comparison:")
print(results_df)
```

Our business goal is to predict the loan buyers and it would be costly to the company to miss classify those clients who are likely to buy a loan. For this purpose, it is crucial to minimize false negatives, which can be identified by maximizing Recall. 

Gradient Boosting and XGBoost models are the models with the highest Recall and the primary candidates for best performing models. On top of that, Gradient Boosting has slightly better Accuracy, Precision and F1 Score than XGBoost. 

Random Forest is also a strong candidate, since it has the highest F1 Score and Precision. In this business case, false positives are not a high risk. But if they were, then Random Forest would be a better option with overall best performance.  


```python
# Function to plot confusion matrix
def plot_confusion_matrix(cm, model_name):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix: {model_name}")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

# Plot confusion matrices for all models
for model_name, cm in confusion_matrices.items():
    plot_confusion_matrix(cm, model_name)
```

### Model Refinement

#### Gradient Boosting

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, recall_score

# Set recall metric as our evaluation and tuning metric
recall_scorer = make_scorer(recall_score)

param_grid = {
    "n_estimators": [100, 200, 300],  # Number of boosting stages
    "learning_rate": [0.01, 0.05, 0.1],  # Shrinkage factor
    "max_depth": [3, 5, 7]  # Maximum tree depth
}

gb_model = GradientBoostingClassifier(random_state=35)

# Perform grid search using recall as the evaluation metric
grid_search = GridSearchCV(
    estimator=gb_model,
    param_grid=param_grid,
    scoring=recall_scorer,  
    cv=5, 
    n_jobs=-1 
)



# Fit the grid search to the training data
grid_search.fit(X_train_balanced, y_train_balanced)

# Output the best parameters and score
print("Best Parameters (Recall):", grid_search.best_params_)
print("Best Cross-Validated Recall:", grid_search.best_score_)
```

Best hyperparameters for Gradient Boosting are: 

    - learning rate: 0.1 
    - max_depth: 5
    - n_estimators: 300

#### Retrain Gradient Boosting

```python
# Train the final Gradient Boosting model with the best parameters
final_gb_model = GradientBoostingClassifier(
    learning_rate=0.1,
    max_depth=5,
    n_estimators=300,
    random_state=35  
)
final_gb_model.fit(X_train_balanced, y_train_balanced)

# Evaluate
y_pred = final_gb_model.predict(X_test_transformed)
y_prob = final_gb_model.predict_proba(X_test_transformed)[:, 1]

# Recall score
test_recall = recall_score(y_test, y_pred)
print(f"Test Recall: {test_recall:.4f}")

# ROC-AUC score
roc_auc = roc_auc_score(y_test, y_prob)
print(f"Test ROC-AUC Score: {roc_auc:.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix: Final Gradient Boosting Model")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

#### XGBoost

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Define the parameter grid for XGBoost
param_grid_xgb = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [3, 5, 7],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}

# Initialize XGBoost
# Initialize XGBoost
xgb_model = XGBClassifier(
    eval_metric="logloss",  # Removed use_label_encoder
    random_state=35
)

# Perform GridSearchCV
grid_search_xgb = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid_xgb,
    scoring="recall",  # Focus on Recall
    cv=5,
    n_jobs=-1
)

# Fit the grid search to the training data
grid_search_xgb.fit(X_train_balanced, y_train_balanced)

# Output the best parameters and score
print("Best Parameters (XGBoost):", grid_search_xgb.best_params_)
print("Best Cross-Validated Recall (XGBoost):", grid_search_xgb.best_score_)
```

Best hyperparameters for XGBoost are: 

    - learning rate: 0.1 
    - max_depth: 5
    - n_estimators: 300
    - subsample: 1.0

#### Refine XGBoost

```python
final_xgb_model = XGBClassifier(
    learning_rate=0.1,
    max_depth=5,
    n_estimators=300,
    subsample=1.0,
    colsample_bytree=1.0,
    random_state=35,
    eval_metric="logloss"
)

final_xgb_model.fit(X_train_balanced, y_train_balanced)

# Evaluate
y_pred = final_xgb_model.predict(X_test_transformed)
y_prob = final_xgb_model.predict_proba(X_test_transformed)[:, 1]

# Recall score
test_recall = recall_score(y_test, y_pred)
print(f"Test Recall: {test_recall:.4f}")

# ROC-AUC score
roc_auc = roc_auc_score(y_test, y_prob)
print(f"Test ROC-AUC Score: {roc_auc:.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix: Final XGBoost Model")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

The refined model has a lower recall score in the test data, indicating overfitting of the training data. Pretuned XGBoost model generalizes better. 

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the model
final_gb_model = GradientBoostingClassifier(
    learning_rate=0.1,
    max_depth=5,
    n_estimators=300,
    random_state=35
)

# Perform cross-validation
gb_cv_scores = cross_val_score(
    estimator=final_gb_model,
    X=X_train_balanced,
    y=y_train_balanced,
    cv=5, 
    scoring="recall"
)

# Output the results
print("Gradient Boosting Cross-Validation Scores (Recall):", gb_cv_scores)
print("Mean Cross-Validated Recall:", gb_cv_scores.mean())


```

The refined model has consistently high recall scores in the 5 folds of subsets with high mean and low std. This ensures the model performs with high efficiency and reliability.

Gradient Boosting (GB) after tuning for best paramters have these advantages over other models (initial and tuned XGBoost): 

1. Achieves a higher recall score than initial model, demonstrating successful optimization for minimizing False Negatives.
2. Shows better alignment between training (cross-validation) and test performance. So it generalizes more effectively than XGBoost in this scenario.
3. Strikes a balance between fitting and predictive power, making it the more reliable model for production.
